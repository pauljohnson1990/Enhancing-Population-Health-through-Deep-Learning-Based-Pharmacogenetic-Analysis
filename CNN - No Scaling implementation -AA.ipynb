# Hyperparameter Tuning

```python
x_train,x_val,y_train,y_val=train_test_split(X,ynew,test_size = 0.2,random_state=76)
```

```python
# Define the objective function for Optuna
def objective(trial):
    # Define the hyperparameters to optimize
    learning_rate = trial.suggest_float("learning_rate", 1e-4, 1e-2, log=True)
    batch_size = trial.suggest_categorical("batch_size", [16, 32, 64])
    filter_size1 = trial.suggest_categorical("filter_size1", [2, 3, 4])
    stride1 = trial.suggest_categorical("stride1", [1, 2])
    filter_size2 = trial.suggest_categorical("filter_size2", [2, 3, 4])
    stride2 = trial.suggest_categorical("stride2", [1, 2])
    filter_size3 = trial.suggest_categorical("filter_size3", [2, 3, 4])
    stride3 = trial.suggest_categorical("stride3", [1, 2])
    padding = trial.suggest_categorical('padding', ['valid', 'same'])

    # Create the model
    model = models.Sequential()
    model.add(layers.Conv2D(64, (filter_size1, filter_size1), strides=(stride1, stride1), activation='relu',padding=padding ,input_shape=(24, 24, 1)))
#     model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(32, (filter_size2, filter_size2), strides=(stride2, stride2),padding=padding,activation='relu'))
#     model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(32, (filter_size3, filter_size3), strides=(stride3, stride3),padding=padding, activation='relu'))
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1, activation='linear'))

    # Compile the model with the chosen hyperparameters
    model.compile(optimizer='adam', loss='mae', metrics=['mse'])

    # Train the model
    history = model.fit(x_train, y_train, epochs=100, batch_size=batch_size, verbose=0)

    # Evaluate the model on the validation set
    loss = model.evaluate(x_val, y_val)[0]

    return loss
```

```python
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=100)

# Get the best hyperparameters and loss from the study
best_params = study.best_params
best_loss = study.best_value

# Print the best hyperparameters and loss
print("Best Hyperparameters: ", best_params)
print("Best Loss: ", best_loss)
```

```python
best_learning_rate = study.best_params["learning_rate"]
best_batch_size = study.best_params["batch_size"]
best_filter_size1 = study.best_params["filter_size1"]
best_stride1 = study.best_params["stride1"]
best_filter_size2 = study.best_params["filter_size2"]
best_stride2 = study.best_params["stride2"]
best_filter_size3 = study.best_params["filter_size3"]
best_stride3 = study.best_params["stride3"]
best_padding = study.best_params["padding"]

best_model = models.Sequential()
best_model.add(layers.Conv2D(64, (best_filter_size1, best_filter_size1), strides=(best_stride1, best_stride1), activation='relu',padding=best_padding, input_shape=(24, 24, 1)))
# best_model.add(layers.MaxPooling2D((2, 2)))
best_model.add(layers.Conv2D(32, (best_filter_size2, best_filter_size2), strides=(best_stride2, best_stride2),padding=best_padding, activation='relu'))
# best_model.add(layers.MaxPooling2D((2, 2)))
best_model.add(layers.Conv2D(32, (best_filter_size3, best_filter_size3), strides=(best_stride3, best_stride3),padding=best_padding, activation='relu'))
best_model.add(layers.Flatten())
best_model.add(layers.Dense(64, activation='relu'))
best_model.add(layers.Dense(1, activation='linear'))

best_model.compile(optimizer='adam', loss='mae', metrics=['mse'])

```

```python
best_model.fit(x_train, y_train, epochs=100, batch_size=best_batch_size, verbose=0)
```

```python
test_loss = best_model.evaluate(x_val, y_val)[0]
print("Test Loss: ", test_loss)
```
